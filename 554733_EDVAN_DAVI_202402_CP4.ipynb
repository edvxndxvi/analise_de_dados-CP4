```python
# Leitura do Dataset
import pandas as pd

df_titanic = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")
df_titanic.head()
```

```python
# Quantidade de linhas e colunas
linhas, colunas = df_titanic.shape
print(f"Há {linhas} linhas e {colunas} colunas no Dataset.")
```

```python
# Dados Faltantes
dados_faltantes = df_titanic.isnull().sum().sum()
print(f"Total de dados faltantes: {dados_faltantes}")
```

```python
# Métricas descritivas para analisar o conjunto de dados
dados = df_titanic.select_dtypes(include='number')
media = dados.values.mean()
mediana = dados.median()
moda = dados.mode().iloc[0]
variancia = dados.var()
desvio_padrao = dados.std()
print(f"Moda dos dados: {moda}")
print(f"Variância dos dados: {variancia}")
print(f"Média dos dados: {media}")
print(f"Mediana dos dados: {mediana}")
print(f"Desvio padrão dos dados: {desvio_padrao}")
```

```python
# Moda, a média, a mediana, a variância e o desvio padrão da colunas Age
media_age = df_titanic['Age'].mean()
mediana_age = df_titanic['Age'].median()
moda_age = df_titanic['Age'].mode().iloc[0]
variancia_age = df_titanic['Age'].var()
desvio_padrao_age = df_titanic['Age'].std()
print(f"Moda da coluna Age: {moda_age}")
print(f"Variância da coluna Age: {variancia_age}")
print(f"Média da coluna Age: {media_age}")
print(f"Mediana da coluna Age: {mediana_age}")
print(f"Desvio padrão da coluna Age: {desvio_padrao_age}")
```

```python
# Moda, a média, a mediana, a variância e o desvio padrão da colunas Fare
media_fare = df_titanic['Fare'].mean()
mediana_fare = df_titanic['Fare'].median()
moda_fare = df_titanic['Fare'].mode().iloc[0]
variancia_fare = df_titanic['Fare'].var()
desvio_padrao_fare = df_titanic['Fare'].std()
print(f"Moda da coluna Fare: {moda_fare}")
print(f"Variância da coluna Fare: {variancia_fare}")
print(f"Média da coluna Fare: {media_fare}")
print(f"Mediana da coluna Fare: {mediana_fare}")
print(f"Desvio padrão da coluna Fare: {desvio_padrao_fare}")
```

```python
# Quantidade de embarques de cada porto presente no dataset
embarques_por_porto = df_titanic['Embarked'].value_counts()
print("Quantidade de embarques por porto:")
print(embarques_por_porto)
```

```python
# Proporção de sobreviventes
proporcao_sobreviventes = df_titanic['Survived'].value_counts(normalize=True)
print("Proporção de sobreviventes:")
print(proporcao_sobreviventes)
```

```python
# Passageiros com a menor e a maior idade presentes nos registros
menor_idade = df_titanic['Age'].min()
maior_idade = df_titanic['Age'].max()

passageiro_menor_idade = df_titanic[df_titanic['Age'] == menor_idade][['Name', 'Age']].iloc[0]

passageiro_maior_idade = df_titanic[df_titanic['Age'] == maior_idade][['Name', 'Age']].iloc[0]

print("Passageiro com a menor idade:")
print(passageiro_menor_idade)

print("\nPassageiro com a maior idade:")
print(passageiro_maior_idade)
```

 ## Gráfico de Sobreviventes por Classe
 O gráfico mostra que passageiros da primeira classe tiveram uma taxa de sobrevivência maior do que os da terceira classe, que teve a maior quantidade de não sobreviventes. Isso indica que a classe de embarque pode ter influenciado as chances de sobrevivência.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Quantidade de sobreviventes e não sobreviventes de acordo com classe de embarque.
plt.figure(figsize=(10, 6))
sns.countplot(data=df_titanic, x='Pclass', hue='Survived', palette='Set2')
plt.title('Quantidade de Sobreviventes e Não Sobreviventes por Classe de Embarque')
plt.xlabel('Classe de Embarque')
plt.ylabel('Quantidade')
plt.legend(title='Sobrevivente', labels=['Não Sobreviveu', 'Sobreviveu'])
plt.show()
```

## Gráfico de relação entre idade e valor pago na passagem
Grande maioria das passagens mais caras foram das mulheres e a maioria das mais baratas foram de homens.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_titanic, x='Age', y='Fare', hue='Sex', palette='viridis', s=100)

plt.title('Relação entre Idade e Valor Pago na Passagem por Sexo')
plt.xlabel('Idade')
plt.ylabel('Valor Pago (Fare)')
plt.legend(title='Sexo')

plt.show()
```

## Gráfico de relação entre idade e valor pago na passagem

O gráfico mostra que a taxa de sobrevivência é maior conforme mais cara a passagem, mostrando também que houve uma taxa de sobrevivência muito grande para pessoas com a idade dos 18, 35 e 36 anos.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_titanic, x='Age', y='Fare', hue='Survived', palette='viridis', s=100)
plt.title('Relação entre Idade e Valor Pago na Passagem de Acordo com o Survived')
plt.xlabel('Idade')
plt.ylabel('Valor Pago')
plt.show()
```

## Grafico de relaçao entre idade e valor pago na passagem
O gráfico mostra que o preço pago para o embarque nos portos Q e S foram muito mais baratos que no porto C, causando uma grande difrença entre Idade e Valor pago na passagem

```python
grupo = df_titanic.groupby('Embarked')[['Age', 'Fare']].mean()
fig, ax1 = plt.subplots(figsize=(10, 6))
grupo['Age'].plot(kind='bar', color='darkcyan', ax=ax1, position=0, width=0.4)
ax2 = ax1.twinx()
grupo['Fare'].plot(kind='bar', color='crimson', ax=ax2, position=1, width=0.4)
ax1.set_title('Média de Idade e Valor Pago por Porto de Embarque (Embarked)')
ax1.set_xlabel('Porto de Embarque')
ax1.set_ylabel('Média de Idade', color='crimson')
ax2.set_ylabel('Média do Valor Pago (Fare)', color='darkcyan')
plt.show()
```

## Grafico de relaçao entre idade e valor pago na passagem
O gráfico mostra uma diferença exorbitante de valor entre a 2ª e 3ª classe comparados a 1ª

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_titanic, x='Age', y='Fare', hue='Pclass', palette='coolwarm', s=100)

plt.title('Relação entre Idade e Valor Pago na Passagem por Classe de Embarque')
plt.xlabel('Idade')
plt.ylabel('Valor Pago (Fare)')
plt.legend(title='Classe de Embarque')

plt.show()
```

## Frequência da idade dos Passageiros
O gráfico mostra que grande parte das pessoas presentes tinham entre 20 e 40 anos

## Frequencia sobre o valor pago nas passagens

```python
# Frequência de Fare
plt.figure(figsize=(10, 6))
sns.histplot(df_titanic['Fare'].dropna(), bins=30, kde=True, color='lightcoral')

plt.title('Distribuição do Valor Pago na Passagem')
plt.xlabel('Valor Pago (Fare)')
plt.ylabel('Frequência')

plt.show()
```
